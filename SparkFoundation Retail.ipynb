```python
plt.figure(figsize=(16,6))
plt.bar('Sub-Category','Category',data=df)
plt.title('Category and sub-category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=50)
plt.show()
```

```python
ggplot(df,aes(x='Category',fill='Sub-Category'))+geom_bar(stat='count')
```

```python
plt.figure(figsize=(8,12))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('')
plt.bar(df['State'],df['Sales'],color='Red',linewidth=1)
plt.xlabel('State',fontsize=15)
plt.ylabel('Sales',fontsize=15)
plt.xticks(fontsize=12,rotation=90)
plt.yticks(fontsize=12)
```

```python
df.plot.scatter(x='Category',y='Profit')
```

```python
df.hist(bins=50,figsize=(20,18))
plt.show()
```

```python
sns.pairplot(df)
```

```python
df=df.loc[(df['Segment']=='Consumer')&(df['Discount']==0.1)]
df['Profit'].plot.hist(bins=40)
```

```python
df=df.loc[(df['Segment']=='Corporate')&(df['Discount']==0.2)]
df['Profit'].plot.hist(bins=40)
```

```python
df=df.loc[(df['Segment']=='Corporate')&(df['Discount']==0.8)]
df['Profit'].plot.hist(bins=40)
```

```python
sns.lineplot('Discount','Profit',data=df,label="Discount")
plt.show()
```

######  Profit is negative for discount 0.3 and above

```python
df['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(df['State'])
plt.xticks(rotation=90)
plt.show()
```

<h1>California and New York has the highest number of buyers

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','blue'])
```

###### profit and sales are maximum in consumer and minimum in home office

```python
df.groupby('State')[['Sales','Profit']].sum().plot.bar(color=['green','red'],figsize=(8,10))
```

###### highest sales are in california,New York and lowest/ loss is in Texas , pennsylvania and ohio.

```python
df.groupby('Category')[['Sales','Profit']].sum().plot.bar(color=['blue','orange'],figsize=(8,8))
```

###### In category technology has more profit than office supplies and furniture and furniture has the lowest profit more work can be done in furniture to make more profit

```python
df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False).plot.bar(color=['red','green'],figsize=(8,10))
```

###### Phones have highest sales but copiers have highest profit
###### Tables have huge loss
