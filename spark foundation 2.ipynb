```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
df1=df.rename(columns={'iyear':'year','imonth':'month','iday':'day','eventid':'id','attacktype1_txt':'attacktype','country_txt':'countryname','region_txt':'regionname'})
df1
```

```python

df1=df1.dropna(axis=1,how='all')
df1
```

```python
df2=df1[['id','year','month','day','country','countryname','region','regionname','success','suicide','attacktype1','attacktype1','targtype1','targtype1_txt','gname','weaptype1','weaptype1_txt','gname']]
```

```python
df2
```

```python
print('country with highest terrorist attacks:',df2['countryname'].value_counts().index[0])
print('regions with highest terrorist attacks:',df2['regionname'].value_counts().index[0])
print('maximum type of weapon used for the attacks:',df2['weaptype1_txt'].value_counts().index[0])
print('maximum attacks targeted at:',df2['targtype1_txt'].value_counts().index[0])
print('The month which had maximum number of attacks:',df2['month'].value_counts().index[0])
print('maximum number of attacks is in the year:',df2['year'].value_counts().index[0])
```

```python
plt.subplots(figsize=(10,8))
sns.countplot('year',data=df2)
plt.xticks(rotation=90)
plt.title('Number of attacks from 1970 to 2017')
plt.show()
```

###### 2014 had the maximum number of terreriost attack

```python
plt.subplots(figsize=(10,8))
sns.countplot('regionname',data=df2)
plt.title('Region with most and least attacks')
plt.xticks(rotation=80)
plt.show
```

###### Middle East and North Africa had the highest number of attacks and Australasia&oceania had the least number of attacks

```python
plt.subplots(figsize=(10,8))
sns.barplot(df2['countryname'].value_counts()[:10].index,df2['countryname'].value_counts()[:10].values,palette='rainbow')
plt.xticks(rotation=45)
plt.title('Country with highest and least number of attacks')
plt.show()
```

###### Iraq has the highest number of attacks and turkey has the least number of attacks

```python
plt.subplots(figsize=(10,8))
sns.countplot(df2['weaptype1_txt'])
plt.xticks(rotation=90)
plt.title('Maximum weapon used for attacks')
plt.show()
```

###### Explosive is the highest number of weapon used for attacks

```python
plt.figure(figsize=(10,8))
df2['countryname'].value_counts()[:10].plot.pie(autopct='%1.1f%%')
plt.show()
```
